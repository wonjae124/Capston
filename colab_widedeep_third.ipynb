```python
import pandas as pd
import torch
from sklearn.preprocessing import LabelEncoder
import numpy as np
from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split

from pytorch_widedeep import Trainer
from pytorch_widedeep.preprocessing import WidePreprocessor, TabPreprocessor
from pytorch_widedeep.models import Wide, TabMlp, WideDeep
from pytorch_widedeep.metrics import Accuracy
from pytorch_widedeep.datasets import load_adult

dataset = pd.read_csv('/content/drive/MyDrive/Wide_deep/aac_intakes_outcomes.csv')
#TODO 메인에서 포크를 시작하는 법 모르므로 해결해야 한다.


dataset = dataset.dropna(axis=0) # 결측치 제거 안 해주면 오류 발생

dataset["Return_to_owner_label"] = (dataset["outcome_type"].apply(lambda x: 'Return_to_Owner' in x)).astype(int)
dataset_train, dataset_test = train_test_split(dataset, test_size = 0.2, stratify = dataset.Return_to_owner_label)
#dataset_train, dataset_test = train_test_split(dataset, test_size = 0.2, stratify = dataset.outcome_type)
```
